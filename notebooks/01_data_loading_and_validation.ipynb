```python
import pandas as pd
import numpy as np
import os
```

```python
# Official UCI dataset URL
DATA_URL = (
    "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/"
    "default%20of%20credit%20card%20clients.xls"
)

print("Dataset source:", DATA_URL)
```

```python
df = pd.read_excel(DATA_URL, header=1)

print("Dataset shape:", df.shape)
df.head()
```

```python
df.rename(
    columns={"default payment next month": "target"},
    inplace=True
)

df["target"].value_counts()
```

```python
# df.isnull().sum()
os.makedirs("data/processed", exist_ok=True)
df.to_csv("data/processed/credit_data.csv", index=False)
print("Processed dataset saved.")
```
